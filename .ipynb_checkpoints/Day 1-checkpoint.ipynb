## count the number of times a depth measurement increases from the previous measurement

In this example, there are 7 measurements that are larger than the previous measurement.

How many measurements are larger than the previous measurement?

199 (N/A - no previous measurement)
200 (increased)
208 (increased)
210 (increased)
200 (decreased)
207 (increased)
240 (increased)
269 (increased)
260 (decreased)
263 (increased)

input:

```python
with open('day1_input.txt') as f:
    contents = f.read()
    print(contents)
```

```python
lines = []
with open('day1_input.txt') as f:
    lines = f.readlines()
for i in range(0,len(lines)):
    lines[i]=lines[i].replace('\n', '')
    #convert str to int
    lines[i]=int(lines[i])
print(lines)
```

```python
count =0
for i in range(1,len(lines)):
    if lines[i]>lines[i-1]:
        print(str(lines[i])+" greater than "+str(lines[i-1]))
        count+=1
        
```

```python
count
```

Part 2

```python
test_input=[199,200,208,210,200,207,240,269,260,263]
```

```python
count =0
part2_sum=[]
for i in range(0,len(test_input)-2):
    sum_test=test_input[i]+test_input[i+1]+test_input[i+2]
    part2_sum.append(sum_test)

for i in range(1,len(part2_sum)):  
    if part2_sum[i]>part2_sum[i-1]:
        print(str(part2_sum[i])+":increased")
        count+=1
    if  part2_sum[i]<=part2_sum[i-1]:
        print(str(part2_sum[i])+":null")
print(count)

```
